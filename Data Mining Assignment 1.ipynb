```python
all_columns = list(train_data.columns)
print('all_columns:\n {}'.format(all_columns))

categorical_columns = list(train_data.select_dtypes(include=['object']).columns)
print('Categorical columns:\n {}'.format(categorical_columns))

numerical_columns = list(train_data.select_dtypes(include=['int64', 'float64']).columns)
print('Numerical columns:\n {}'.format(numerical_columns))
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
encoded_adult_data = train_data
for i in categorical_columns:
    encoded_adult_data[i] = label_encoder.fit_transform(train_data[i])
encoded_adult_data[categorical_columns].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

scaled_encoded_adult_data = pd.DataFrame()

column_values = encoded_adult_data.columns.values
column_values = column_values[:-1]
print(column_values[-1])

scaled_values = min_max_scaler.fit_transform(encoded_adult_data[column_values])

for i in range(len(column_values)):
    scaled_encoded_adult_data[column_values[i]] = scaled_values[:,i]
    
scaled_encoded_adult_data['income'] = encoded_adult_data['income']
scaled_encoded_adult_data.sample(10)
```

```python
train_data.shape
```

```python
new_columns = numerical_columns.copy()
new_columns.remove('capital-gain') 
new_columns.remove('capital-loss') 
new_columns
```

```python
train_data.tail
```

```python
def outlier_detector(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn,[25,75])
    IQR = Q3 - Q1
    lower_bound = Q1-(1.5*IQR)
    upper_bound = Q3+(1.5*IQR)
    return lower_bound,upper_bound

lowerbound, upperbound = outlier_detector(scaled_encoded_adult_data['age'])
lowerbound, upperbound

scaled_encoded_adult_data[(scaled_encoded_adult_data.age < lowerbound) | (scaled_encoded_adult_data.age > upperbound)]

```

```python
treated_scaled_encoded_adult_data = scaled_encoded_adult_data.copy()
for i in new_columns:
    lowerbound, upperbound = outlier_detector(treated_scaled_encoded_adult_data[i])
    median = treated_scaled_encoded_adult_data[i].median()
    treated_scaled_encoded_adult_data[i] = np.where((treated_scaled_encoded_adult_data[i] < lowerbound) | 
                                                     (treated_scaled_encoded_adult_data[i] > upperbound), 
                                                     median, 
                                                     treated_scaled_encoded_adult_data[i])
    print('{}: numbers of outliers: {}'.format(i,len(treated_scaled_encoded_adult_data[
        (treated_scaled_encoded_adult_data[i] < lowerbound) |
        (treated_scaled_encoded_adult_data[i] > upperbound)])))
```

```python
treated_scaled_encoded_adult_data.tail
```

```python
print(all_columns)

features = all_columns[:-1]
target = treated_scaled_encoded_adult_data['income']
print(features)
print(treated_scaled_encoded_adult_data.shape)
```

```python
feature_df = treated_scaled_encoded_adult_data[features]
print(target.head())
feature_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(feature_df, target, test_size=0.2)
print(x_train.shape,y_train.shape, x_test.shape, y_test.shape)
```

**Model Training**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import time

# Define the parameter grid
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}

# Initialize the SVC model
svc = SVC(kernel='rbf')

# Use GridSearchCV to find the best parameters
start_time = time.time()
grid = GridSearchCV(svc, param_grid, cv=5)
grid.fit(x_train, y_train)
end_time = time.time()

# Get the best parameters found by GridSearchCV
best_C = grid.best_params_['C']
best_gamma = grid.best_params_['gamma']

# Train the SVC model with the best parameters
svc = SVC(kernel='rbf', C=best_C, gamma=best_gamma)
start_time2 = time.time()
svc.fit(x_train, y_train)
end_time2 = time.time()

# Get the training and testing scores
svc_train_score = svc.score(x_train, y_train)
svc_test_score = svc.score(x_test, y_test)

# Print out the training and testing scores, as well as the time taken for GridSearchCV and training
print("Training score:", svc_train_score)
print("Testing score:", svc_test_score)
print("Time taken for GridSearchCV:", end_time - start_time)
print("Time taken for training:", end_time2 - start_time2)

```

```python
from sklearn.metrics import mean_squared_error

svc_prediction = svc.predict(x_test)

svc_mse = mean_squared_error(y_test, svc_prediction)
svc_rmse = np.sqrt(svc_mse)

print('MSE: {}\nRMSE: {}'.format(svc_mse, svc_rmse))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import time

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50, 60],
    'min_samples_split': [2, 5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': [None, 'sqrt', 'log2'],
    'splitter': ['best', 'random']
}


dtree_classifier = DecisionTreeClassifier()
grid_search = GridSearchCV(dtree_classifier, param_grid, cv=5)

start_time = time.time()
grid_search.fit(x_train, y_train)
end_time = time.time()
grid_time = end_time - start_time

dtree_classifier = grid_search.best_estimator_

start_time = time.time()
dtree_classifier.fit(x_train, y_train)
end_time = time.time()
train_time = end_time - start_time

dtree_train_score = dtree_classifier.score(x_train, y_train)
dtree_test_score = dtree_classifier.score(x_test, y_test)

print("Decision Tree Train Score: ", dtree_train_score)
print("Decision Tree Test Score: ", dtree_test_score)
print("Time taken for GridSearchCV: {:.6f}s".format(grid_time))
print("Time taken for training: {:.6f}s".format(train_time))
```

```python
dtree_prediction = dtree_classifier.predict(x_test)

dtree_mse = mean_squared_error(y_test, dtree_prediction)
dtree_rmse = np.sqrt(dtree_mse)

print('MSE: {}\nRMSE: {}'.format(dtree_mse, dtree_rmse))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Compute predicted probabilities for SVC
svc_probs = svc.decision_function(x_test)
svc_fpr, svc_tpr, _ = roc_curve(y_test, svc_probs)
svc_roc_auc = auc(svc_fpr, svc_tpr)

# Compute predicted probabilities for decision tree
dtree_probs = dtree_classifier.predict_proba(x_test)[:, 1]
dtree_fpr, dtree_tpr, _ = roc_curve(y_test, dtree_probs)
dtree_roc_auc = auc(dtree_fpr, dtree_tpr)

print('SVC ROC AUC: %.3f' % svc_roc_auc)
print('Decision Tree ROC AUC: %.3f' % dtree_roc_auc)


# Plot the ROC curves
plt.plot(svc_fpr, svc_tpr, color='blue', label='SVC ROC (AUC = %0.2f)' % svc_roc_auc)
plt.plot(dtree_fpr, dtree_tpr, color='orange', label='Decision Tree ROC (AUC = %0.2f)' % dtree_roc_auc)
plt.plot([0, 1], [0, 1], color='black', linestyle='--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Compute predictions for SVC
svc_train_pred = svc.predict(x_train)
svc_test_pred = svc.predict(x_test)

# Compute predictions for decision tree
dtree_train_pred = dtree_classifier.predict(x_train)
dtree_test_pred = dtree_classifier.predict(x_test)

# Compute metrics for SVC
svc_train_accuracy = accuracy_score(y_train, svc_train_pred)
svc_test_accuracy = accuracy_score(y_test, svc_test_pred)
svc_train_precision = precision_score(y_train, svc_train_pred)
svc_test_precision = precision_score(y_test, svc_test_pred)
svc_train_recall = recall_score(y_train, svc_train_pred)
svc_test_recall = recall_score(y_test, svc_test_pred)
svc_train_f1 = f1_score(y_train, svc_train_pred)
svc_test_f1 = f1_score(y_test, svc_test_pred)

# Compute metrics for decision tree
dtree_train_accuracy = accuracy_score(y_train, dtree_train_pred)
dtree_test_accuracy = accuracy_score(y_test, dtree_test_pred)
dtree_train_precision = precision_score(y_train, dtree_train_pred)
dtree_test_precision = precision_score(y_test, dtree_test_pred)
dtree_train_recall = recall_score(y_train, dtree_train_pred)
dtree_test_recall = recall_score(y_test, dtree_test_pred)
dtree_train_f1 = f1_score(y_train, dtree_train_pred)
dtree_test_f1 = f1_score(y_test, dtree_test_pred)

# Print metrics for SVC
print("SVC Training Accuracy: ", svc_train_accuracy)
print("SVC Testing Accuracy: ", svc_test_accuracy)
print("SVC Training Precision: ", svc_train_precision)
print("SVC Testing Precision: ", svc_test_precision)
print("SVC Training Recall: ", svc_train_recall)
print("SVC Testing Recall: ", svc_test_recall)
print("SVC Training F1 Score: ", svc_train_f1)
print("SVC Testing F1 Score: ", svc_test_f1)

# Print metrics for decision tree
print("Decision Tree Training Accuracy: ", dtree_train_accuracy)
print("Decision Tree Testing Accuracy: ", dtree_test_accuracy)
print("Decision Tree Training Precision: ", dtree_train_precision)
print("Decision Tree Testing Precision: ", dtree_test_precision)
print("Decision Tree Training Recall: ", dtree_train_recall)
print("Decision Tree Testing Recall: ", dtree_test_recall)
print("Decision Tree Training F1 Score: ", dtree_train_f1)
print("Decision Tree Testing F1 Score: ", dtree_test_f1)
```
